```python
boston.data.shape
```

```python
print(boston.feature_names) # Print columns name
```

```python
# Print description of Boston housing data set
print(boston.DESCR)
```

```python
bos = pd.DataFrame(boston.data)
bos.head()
```

```python
bos.columns = boston.feature_names
bos.head()
```

```python
bos["PRICE"] = boston.target
bos.head()
```

```python
bos.describe()
```

```python
plt.scatter(bos.CRIM, bos.PRICE)
plt.xlabel("Per capita crime rate by town (CRIM)")
plt.ylabel("Housing Price")
plt.title("Relationship between CRIM and Price")
```

```python
plt.scatter(bos.RM, bos.PRICE)
plt.xlabel("Avg number of rooms per dwelling (RM)")
plt.ylabel("Housing Price")
plt.title("Relationship between RM and Price")
```

```python
rows = 2
cols = 7
fig, ax = plt.subplots(nrows= rows, ncols= cols, figsize = (16,4))
col = bos.columns
index = 0
for i in range(rows):
    for j in range(cols):
        sns.distplot(bos[col[index]], ax = ax[i][j])
        index = index + 1
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA

test = bos["PRICE"][-20:]
train = bos["PRICE"][:-20]

model = ARIMA(train, order=(2,0,1))
model_fit = model.fit(disp=0)
#summary = model_fit.summary()
predictions = model_fit.predict(len(train), len(train)+len(test)-1)
pd.DataFrame({'observed':test, 'predicted':predictions})
```

```python
y = bos["PRICE"]
f = bos["TAX"]
r2 = r2_score(y, f)
print(r2)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(bos[['LSTAT']], bos['PRICE'], random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
LRmodel= LinearRegression()
LRmodel.fit(X_train,y_train)
Ypred=LRmodel.predict(X_test)

plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, Ypred, color='red', linewidth=1)
plt.xticks(())
plt.yticks(())
plt.xlabel('LSTAT')
plt.ylabel('PRICE')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import math
actual = bos[['LSTAT']]
predicted = bos['PRICE']

mse = mean_squared_error(actual, predicted)

rmse = math.sqrt(mse)
print(mse)
print(rmse)
```

```python
from sklearn.metrics import mean_squared_error, r2_score  
print("MSE: {:.2f}".format(mean_squared_error(y_test, Ypred)))
```
